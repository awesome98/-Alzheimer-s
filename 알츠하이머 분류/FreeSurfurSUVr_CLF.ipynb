## 랜덤 포레스트 단일 시행

### 상관계수가 높은 feature만 시행

```python
from sklearn.model_selection import train_test_split

X = df_over_corr.drop('Diagnosis code', axis=1)
y = df_over_corr['Diagnosis code']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score # 정확도 함수

rnd_clf = RandomForestClassifier()
rnd_clf.fit(x_train, y_train)
```

```python
predict = rnd_clf.predict(x_test)
print(accuracy_score(y_test,predict))
```

### 하이퍼 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = { 'n_estimators' : [50, 100, 150, 170],
           'max_depth' : [None,6,7,8,9,10,15],
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(criterion='entropy', n_jobs = -1, random_state=42)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 5, n_jobs = -1)
grid_cv.fit(x_train, y_train)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
rnd_clf_1 = RandomForestClassifier(max_depth=6, n_estimators=150, n_jobs=-1)
rnd_clf_1.fit(x_train, y_train)
```

```python
predict1 = rnd_clf_1.predict(x_test)
print(accuracy_score(y_test,predict1))
```

## LDA(선형 판별 분석) 시행

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda_clf = LinearDiscriminantAnalysis()
lda_clf.fit(x_train, y_train)
```

```python
lda_clf.transform(x_train).shape
```

```python
predict2 = lda_clf.predict(x_test)
print(accuracy_score(y_test, predict2))
```

```python
param_grid = {
    'solver': ['svd', 'lsqr', 'eigen'],  # solver 옵션
    'shrinkage': [None, 'auto', 0.1, 0.5, 0.9],  # shrinkage 옵션
}

lda_clf_1 = LinearDiscriminantAnalysis()

# GridSearchCV를 사용하여 최적의 하이퍼파라미터 찾기
grid_search = GridSearchCV(lda_clf_1, param_grid, cv=5)
grid_search.fit(x_train, y_train)

# 최적의 하이퍼파라미터와 그 때의 평가 점수 출력
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
lda_clf_1.fit(x_train, y_train)
predict3 = lda_clf_1.predict(x_test)
print(accuracy_score(y_test, predict3))
```
