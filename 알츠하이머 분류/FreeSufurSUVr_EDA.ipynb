## 2차원 시각화

```python
from sklearn.preprocessing import StandardScaler 

# 먼저 PCA를 위해 데이터를 표준화해줍니다.
data = df_corr.drop(['Diagnosis code','Onset age'], axis=1)
target = df_corr['Diagnosis code']

scaled_data = StandardScaler().fit_transform(data)
```

```python
from sklearn.decomposition import PCA

# 2D 에 시각화하기 위해 주성분을 2개로 선택합니다.
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
principalDf = pd.DataFrame(data=pca_data, columns = ['principal component 1', 'principal component 2'])
principalDf['Diagnosis code'] = target
```

```python
plt.figure(figsize = (20, 15))

plt.xlabel('Principal Component 1', fontsize = 15)
plt.ylabel('Principal Component 2', fontsize = 15)
plt.title('2 component PCA', fontsize = 40)

sns.scatterplot(x='principal component 1', y='principal component 2', data=principalDf, hue='Diagnosis code', s= 100)
plt.show()
```

## 3차원 시각화

```python
from sklearn.decomposition import PCA

# 3D 에 시각화하기 위해 주성분을 3개로 선택합니다.
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_data)
principalDf = pd.DataFrame(data=pca_data, columns=['principal component 1', 'principal component 2', 'principal component 3'])
principalDf['Diagnosis code'] = target

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('Principal Component 1', fontsize=12)
ax.set_ylabel('Principal Component 2', fontsize=12)
ax.set_zlabel('Principal Component 3', fontsize=12)
ax.set_title('3D PCA', fontsize=16)

scatter = ax.scatter(principalDf['principal component 1'], principalDf['principal component 2'], principalDf['principal component 3'],
                     c=principalDf['Diagnosis code'], cmap='viridis', s=100)

legend1 = ax.legend(*scatter.legend_elements(), title='Diagnosis code')
ax.add_artist(legend1)

plt.show()
```
