#  카테고리 분류 : 0,1,2,3,4 / 5 

```python
df1 = df.copy()
```

```python
df1.head()
```

```python
df1['Diagnosis code'] = df1['Diagnosis code'].apply(lambda x: 1 if x==5 else 0)
df1['Diagnosis code'].unique()
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split

X = df1.drop('Diagnosis code', axis=1)
y = df1['Diagnosis code']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score # 정확도 함수

rnd_clf = RandomForestClassifier()
rnd_clf.fit(x_train, y_train)
```

```python
predict = rnd_clf.predict(x_test)
print(accuracy_score(y_test,predict))
```

## 하이퍼 파라미터 튜닝

```python
from sklearn.model_selection import GridSearchCV

params = { 'n_estimators' : [50, 100, 150, 170],
           'max_depth' : [None,6,7,8,9,10,15],
            }

# RandomForestClassifier 객체 생성 후 GridSearchCV 수행
rf_clf = RandomForestClassifier(criterion='entropy', n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 5, n_jobs = -1)
grid_cv.fit(x_train, y_train)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
rnd_clf_1 = grid_cv.best_estimator_
rnd_clf_1.fit(x_train, y_train)
```

```python
predict1 = rnd_clf_1.predict(x_test)
print(accuracy_score(y_test,predict1))
```

```python
from xgboost import XGBClassifier

xgb_clf = XGBClassifier(n_estimators=500, learning_rate=0.1, max_depth=6)
xgb_clf.fit(x_train, y_train)
```

```python
predict11 = xgb_clf.predict(x_test)
print(accuracy_score(y_test,predict11))
```

```python
# XGBoost 모델 정의
model = XGBClassifier()

# GridSearch를 위한 파라미터 그리드 설정
param_grid = {
    'n_estimators': [50, 100, 150, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# GridSearchCV 설정
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='accuracy',  # 적절한 평가 지표 선택
    cv=5,                # 교차 검증 폴드 수
    verbose=2            # 진행 과정 출력 설정
)

# GridSearch 실행
grid_search.fit(x_train, y_train)

# 최적 파라미터 및 결과 출력
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validated Score:", grid_search.best_score_)
```

```python
xgb_clf_1 = grid_search.best_estimator_
xgb_clf_1.fit(x_train, y_train)
```

```python
predict_xgb = xgb_clf_1.predict(x_test)
print(accuracy_score(y_test,predict_xgb))
```

## 랜덤 포레스트와 XGBoost의 앙상블 모델

```python
from sklearn.ensemble import VotingClassifier

# VotingClassifier 설정
voting_model = VotingClassifier(
    estimators=[('random_forest', rnd_clf_1), ('xgboost', xgb_clf_1)],
    voting='soft'  # Soft Voting 설정
)

# 앙상블 모델 훈련
voting_model.fit(x_train, y_train)

# 테스트 세트 예측
y_pred = voting_model.predict(x_test)

# 정확도 평가
accuracy = accuracy_score(y_test, y_pred)
print("Ensemble Model Test Set Accuracy:", accuracy)
```
